```python
# item_id: book_title
# user_id: rname
# rating: stars_given

# Building a Utility matrix

utility_matrix = all_reviews.pivot_table(values = 'stars_given', index = 'rname', columns = 'book_title') #, fill_value=0)
utility_matrix
```

```python
# Data Normalization

# Normalize user-item matrix
matrix_norm = utility_matrix.subtract(utility_matrix.mean(axis=1), axis = 'rows')
matrix_norm.head()
```

```python
# Identify similar users

# User similarity matrix using Pearson correlation
user_similarity = matrix_norm.T.corr()
user_similarity.head()
```

```python
# User similarity matrix using cosine similarity
user_similarity_cosine = cosine_similarity(matrix_norm.fillna(0))
user_similarity_cosine
```

```python
# Pick a user ID
picked_userid = "7jane"
# Remove picked user ID from the candidate list
user_similarity.drop(index=picked_userid, inplace=True)
# Take a look at the data
user_similarity.head()
```

```python
# Number of similar users
n = 5
# User similarity threashold
user_similarity_threshold = 0.3
# Get top n similar users
similar_users = user_similarity[user_similarity[picked_userid]>user_similarity_threshold][picked_userid].sort_values(ascending=False)[:n]
# Print out top n similar users
print(f'The similar users for user {picked_userid} are', similar_users)
```

```python
# Books that the target user has reviewed
picked_userid_reviewed = matrix_norm[matrix_norm.index == picked_userid].dropna(axis=1, how='all')
picked_userid_reviewed
```

```python
# Books that similar users reviewed. Remove books that none of the similar users have reviewed
similar_user_books = matrix_norm[matrix_norm.index.isin(similar_users.index)].dropna(axis=1, how='all')
similar_user_books
```

```python
# Remove the reviewed book from the books list
similar_user_books.drop(picked_userid_reviewed.columns,axis=1, inplace=True, errors='ignore')
# Take a look at the data
similar_user_books
```

```python
# A dictionary to store item scores
item_score = {}

# Loop through items
for i in similar_user_books.columns:
  # Get the review for book i
  book_review = similar_user_books[i]
  # Create a variable to store the score
  total = 0
  # Create a variable to store the number of scores
  count = 0
  # Loop through similar users
  for u in similar_users.index:
    # If the book has review
    if pd.isna(book_review[u]) == False:
      # Score is the sum of user similarity score multiply by the book review
      score = similar_users[u] * book_review[u]
      # Add the score to the total score for the book so far
      total += score
      # Add 1 to the count
      count +=1
  # Get the average score for the item
  item_score[i] = total / count
# Convert dictionary to pandas dataframe
item_score = pd.DataFrame(item_score.items(), columns=['book', 'book_score'])
    
# Sort the movies by score
ranked_item_score = item_score.sort_values(by='book_score', ascending=False)
# Select top m movies
m = 10
ranked_item_score.head(m)
```

```python
# Predict score

# Average stars_given for the picked user
avg_rating = utility_matrix[utility_matrix.index == picked_userid].T.mean()[picked_userid]
# Print the average book stars for user 7jane
print(f'The average book stars for user {picked_userid} is {avg_rating:.2f}')
```

```python
# Calcuate the predicted rating
ranked_item_score['predicted_stars'] = ranked_item_score['book_score'] + avg_rating
# Take a look at the data
ranked_item_score.head(m)
```
